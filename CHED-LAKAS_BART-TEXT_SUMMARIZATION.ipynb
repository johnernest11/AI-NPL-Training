```python
#Import libraries
import torch
from transformers import BartTokenizer, BartForConditionalGeneration
from transformers import pipeline
```

###  Load the pre-trained BERT model and tokenizer for summarization

```python
model_name = "facebook/bart-large-cnn"
tokenizer = BartTokenizer.from_pretrained(model_name)
model = BartForConditionalGeneration.from_pretrained(model_name)
```

### # Create a text summarization pipeline using BERT

```python
while True:
    # Get user input
    user_input = input("Enter a sentence to summarize (or type '*' to exit): ")

    # Check if the user wants to exit
    if user_input == "*":
        break

    # Tokenize and generate the summary
    inputs = tokenizer(user_input, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = model.generate(inputs["input_ids"], max_length=150, min_length=30, length_penalty=2.0, num_beams=4, early_stopping=True)

    # Decode and print the summary
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    print("Summary:")
    print(summary)
```
